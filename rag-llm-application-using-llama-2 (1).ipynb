```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.core.prompts.prompts import SimpleInputPrompt
from llama_index.llms.huggingface import HuggingFaceLLM
import torch
```

```python
documents=SimpleDirectoryReader("/kaggle/input/attention").load_data()
```

```python
system_prompt= """
You are a Q&A assistant.Your goal is to answer questions as 
accurately as possible based on instruction and context provided
"""

query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
query_wrapper_prompt
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
llm= HuggingFaceLLM(context_window=4096,
                   max_new_tokens=256,
                   generate_kwargs={"temperature":0.3,"do_sample":False},
                   system_prompt=system_prompt,
                   query_wrapper_prompt=query_wrapper_prompt,
                   tokenizer_name="Shorya22/LLaMA-2-7B",
                   model_name="Shorya22/LLaMA-2-7B",
                   device_map="auto",
                   model_kwargs={"torch_dtype":torch.float16,"load_in_8bit":True})
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.legacy.embeddings.langchain import LangchainEmbedding
```

```python
embed_model= LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context= ServiceContext.from_defaults(chunk_size=1024,
                                             llm=llm,
                                             embed_model=embed_model)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response= query_engine.query("can you please explain what is attention?")
```
